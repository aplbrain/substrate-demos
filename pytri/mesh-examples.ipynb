```python
from pytri import pytri
```

```python
p = pytri()
```

```python
obj_data = """
v -5 -5 -5
v -5 -5 5
v -5 5 -5
v -5 5 5
v 5 -5 -5
v 5 -5 5
v 5 5 -5
v 5 5 5
f 1 2 4 3
f 5 6 8 7
f 1 2 6 5
f 3 4 8 7
"""

p.mesh(obj_data)
p.show()
```

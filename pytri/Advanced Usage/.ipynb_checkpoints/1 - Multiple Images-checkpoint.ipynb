# Multiple images

For this usecase, we have a 3D volume of data that we want to view as individual slices.

```python
import numpy as np
from pytri import pytri
```

```python
data = np.random.randint(0, 255, (50, 50, 50)).astype('float')
```

```python
p = pytri()
```

```python
p.show()
```

```python
images = [
    p.imshow(data[i], position={"x": 0, "y": 0, "z": i/5})
    for i in range(len(data))
]
```

```python
for im in images:
    p.toggle_layer(im)
p.toggle_layer(images[0])
```
